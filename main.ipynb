```python
from langchain.vectorstores import Chroma
from langchain_ollama import ChatOllama, OllamaEmbeddings
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
```

```python
embedding_function = OllamaEmbeddings(
    model="nomic-embed-text",
)
```

```python
def alan_turing_bio(question):
    '''
    Takes in a question about Alan Turing and returns the most relevant
    part of his biography using a compressed context retriever.
    '''
    # PART ONE: Load the document
    loader = TextLoader("AlanTuring.txt")
    documents = loader.load()

    # PART TWO: Split into chunks
    text_splitter = CharacterTextSplitter.from_tiktoken_encoder(chunk_size=500, chunk_overlap=50,separator="\n\n")
    docs = text_splitter.split_documents(documents)

    # PART THREE: Embed the chunks and persist to ChromaDB
    embeddings = embedding_function.embed_documents([doc.page_content for doc in docs])

    # Create Chroma DB with precomputed embeddings
    db = Chroma.from_documents(
        docs,
        embedding=embedding_function,
        persist_directory="./Alan_embedding_db",
        ids=[str(i) for i in range(len(docs))],
        embeddings=embeddings
)
    db.persist()

    # PART FOUR: Use ChatOllama and ContextualCompressionRetriever
    llm = ChatOllama(model="gemma3:27b")
    compressor = LLMChainExtractor.from_llm(llm)
    retriever = ContextualCompressionRetriever(base_compressor=compressor, base_retriever=db.as_retriever())

    # Retrieve the most relevant document
    relevant_docs = retriever.get_relevant_documents(question)
    if relevant_docs:
        return relevant_docs[0].page_content
    else:
        return "No relevant information found."
```
